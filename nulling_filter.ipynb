```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
from pylab import mpl

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams['axes.unicode_minus']=False
```

```python
fs = 100
t = np.linspace( 0, 1, fs, endpoint = False )   # 定義時間陣列
x1 = np.cos( 2 * np.pi * 10 * t )               # 原始訊號
x2 = np.cos( 2 * np.pi * 20 * t )               # 干擾訊號
x = x1 + x2
```

```python
b = np.array( [ 1, -2 * np.cos( 2 * np.pi * 20 / fs ), 1 ] )
y = signal.lfilter( b, 1, x )
```

```python
plt.figure( 1 )                                 # 繪圖
plt.plot( t, x )
plt.title(u"輸入訊號")
plt.xlabel( 't (second)' )
plt.ylabel( 'Amplitude' )
plt.axis( [ 0, 1, -2, 2 ] )

plt.show( )
```

```python
plt.figure( 2 )
plt.plot( t, x1, '--', label = u"原始訊號")
plt.plot( t, y, '-' , label = u"輸出訊號")
plt.title(u"FIR濾波(與原始訊號比較)")
plt.xlabel( 't (second)' )
plt.ylabel( 'Amplitude' )
plt.legend(loc = "best")
plt.axis( [ 0, 1, -2, 2 ] )

plt.show( )
```

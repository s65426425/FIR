```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
```

```python
x = np.array( [ 1, 2, 4, 3, 2, 1, 1 ] )
b = np.ones( 3 ) / 3
y = signal.lfilter( b, 1, x )

print( "x =", x )
print( "b =", b )
print( "y =", y )
```

```python
plt.figure(1)                             # 繪圖
plt.plot( x, '--', label = "x" )
plt.plot( y, '-', label = "y" )
plt.xlabel( 't' )
plt.ylabel( 'distance' )
plt.legend(loc = "best")

plt.show()
```
